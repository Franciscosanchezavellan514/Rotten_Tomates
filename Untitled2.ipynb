```python
import pandas as pd
import matplotlib.pyplot as plt

df_movies = pd.read_csv(r"C:\Users\dell_14\Desktop\TAREA ROTEN\Rotten_Tomates\Rotten Tomatoes Movies.csv")

# Calcular promedio
promedio_criticos = df_movies['tomatometer_rating'].mean()
promedio_audiencia = df_movies['audience_rating'].mean()

print(f"Promedio cr√≠ticos: {promedio_criticos:.2f}")
print(f"Promedio audiencia: {promedio_audiencia:.2f}")

# Histograma de diferencia
df_movies['rating_diff'] = df_movies['audience_rating'] - df_movies['tomatometer_rating']

plt.hist(df_movies['rating_diff'].dropna(), bins=30, color='skyblue', edgecolor='black')
plt.title("Distribucion,diferencias entre audiencia y cr√≠ticos")
plt.xlabel("Diferencia (audiencia - cr√≠ticos)")
plt.ylabel("N√∫mero de pel√≠culas")
plt.show()
```
