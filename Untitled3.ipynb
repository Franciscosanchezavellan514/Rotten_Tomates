```python

import pandas as pd
import matplotlib.pyplot as plt

df_movies = pd.read_csv(r"C:\Users\dell_14\Desktop\TAREA ROTEN\Rotten_Tomates\Rotten Tomatoes Movies.csv")

df_genres_exploded = df_movies.assign(
    genre=df_movies['genre'].str.split(',')
).explode('genre')
df_genres_exploded['genre'] = df_genres_exploded['genre'].str.strip()

promedios_por_genero = df_genres_exploded.groupby('genre')['audience_rating'].mean().sort_values(ascending=False)

top_10_generos = promedios_por_genero.head(10)
print(top_10_generos)

top_10_generos.plot.pie(autopct='%1.1f%%', figsize=(8,8), title="Top 10 peliculas con mejor promdio de valoracion")
plt.ylabel("")
plt.show()
```
