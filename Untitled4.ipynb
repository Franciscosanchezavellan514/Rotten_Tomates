```python
import pandas as pd
import matplotlib.pyplot as plt

df_movies = pd.read_csv(r"C:\\Users\\dell_14\\Desktop\\TAREA ROTEN\\Rotten_Tomates\\Rotten Tomatoes Movies.csv")

top_10_directors = df_movies['directors'].value_counts().head(10)
top_10_director_names = top_10_directors.index.tolist()

df_top_10 = df_movies[df_movies['directors'].isin(top_10_director_names)].copy()

promedios = df_top_10.groupby('directors')['tomatometer_rating'].mean().sort_values(ascending=False)

plt.figure(figsize=(10,6))
promedios.plot(kind='bar', color='skyblue')
plt.title("Calificacin Promedio (Tomatometer) de los Top 10 Directores Ms Frecuentes")
plt.xlabel("Calificacin Promedio")
plt.ylabel("Director")
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
