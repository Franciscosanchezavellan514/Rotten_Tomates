```python
import pandas as pd
import csv

file_path = "C:/Users/dell_14/Desktop/Data/Rotten Tomatoes Movies.csv"
df_movies = pd.read_csv(file_path, sep=',', quoting=csv.QUOTE_NONE, on_bad_lines='skip')

print("Primeras 5 filas del dataset:\n")
print(df_movies.head())


print("\nTipos de datos de cada columna:")
print(df_movies.info())  

expected_columns = ['movie_title', 'genre', 'tomatometer_rating', 'audience_rating', 
                    'in_theaters_date', 'on_streaming_date', 'directors', 
                    'tomatometer_status', 'runtime_in_minutes']
missing_columns = [col for col in expected_columns if col not in df_movies.columns]
if missing_columns:
    print(f"Las siguientes columnas están faltando: {missing_columns}")
else:
    print("Todas las columnas principales están presentes.")

df_movies['in_theaters_date'] = pd.to_datetime(df_movies['in_theaters_date'], errors='coerce')

print("\nTipos de datos actualizados:")
print(df_movies.dtypes)

missing_dates = df_movies['in_theaters_date'].isna().sum()
print(f"\nPelículas con fechas no reconocidas (NaT): {missing_dates}")
```
